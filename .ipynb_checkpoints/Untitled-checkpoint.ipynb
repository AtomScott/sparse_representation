# Sparse Representation

## ALOIDataset

ALOI is a color image collection of one-thousand small objects, recorded for scientific purposes. In order to capture the sensory variation in object recordings, we systematically varied viewing angle, illumination angle, and illumination color for each object, and additionally captured wide-baseline stereo images. We recorded over a hundred images of each object, yielding a total of 110,250 images for the collection. See Technical Details for a description of the acquisition setup.

- Illumination Direction is varied in 24 configurations
- Illumination Color is varied in 12 configurations
- Object Viewpoint is varied from 72 directions

Full dataset can be downloaded here: 
https://aloi.science.uva.nl/

```python
from scipy.io import loadmat
X = loadmat('aloi200.mat')["data"]
print(X.shape)

import numpy as np
y = np.asarray([c for _ in range(X.shape[-2]) for c in range(X.shape[-1])])

from einops import rearrange
X = rearrange(X, 'h w view c -> (view c) (h w)')
print(X.shape)
print(y.shape)
```

```python
import matplotlib.pyplot as plt

# Visualize and check labelling
N = 25
plt.imshow(rearrange(X[::20][:N], '(b1 b2) (h w) -> (b1 h) (b2 w)', b1 = int(np.sqrt(N)), h=15))
plt.show()
print(rearrange(y[::20][:N], '(b1 b2) -> b1 b2', b1 = int(np.sqrt(N))))
```

```python
from sklearn.model_selection import train_test_split
from scipy.stats import describe

# Divide train and test samples
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42, stratify=y)

# Check if splits are properly stratified along classes
print(describe(np.unique(y_train, return_counts=True)[1]))
print(describe(np.unique(y_test, return_counts=True)[1]))
```

## Sparse Representation-based Classification (SRC)

2009 TPAMI Robust Face Recognition Via Sparse Representation

```python
from numpy.linalg import norm, pinv
from numpy.testing import assert_almost_equal

# normalise columns of the trainset to have unit L2-norm
X_train = X_train / norm(X_train, ord=2, axis=1, keepdims=True)
assert_almost_equal(norm(X_train, ord=2, axis=1).mean(), 1)

# solve the l1-minimization problem
from scipy.optimize import minimize
```

```python
import optuna
import cvxpy as cp
from tqdm import tqdm

def loss_fn(X, Y, beta):
    return cp.norm2(X @ beta - Y)**2

def regularizer(beta):
    return cp.norm1(beta)

def objective_fn(X, Y, beta, lambd):
    return loss_fn(X, Y, beta) + lambd * regularizer(beta)

def mse(X, Y, beta):
    return (1.0 / X.shape[0]) * loss_fn(X, Y, beta).value


def optuna_objective(trial):
    A = X_train.T
    beta = cp.Variable(len(X_train), nonneg=True)
    lambd = cp.Parameter(nonneg=True)

    lambd.value = trial.suggest_loguniform('lambda',10**-1,10**2)
    print(lambd.value)
    score = 0
    for i, y in tqdm(enumerate(X_test)):
        problem = cp.Problem(cp.Minimize(objective_fn(A, y, beta, lambd)))
        problem.solve()
        
        y_pred = np.argmax([beta.value[np.where(y_train==c)].sum() for c in range(200)])
        score += int(y_pred == y_test[i])
        print(y_pred, y_test[i])
        plt.bar(range(200), [beta.value[np.where(y_train==c)].sum() for c in range(200)])
        plt.show()
    return score / i
   

study = optuna.create_study()
study.optimize(optuna_objective, n_trials=100)
```
